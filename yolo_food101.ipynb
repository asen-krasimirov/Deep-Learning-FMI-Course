The Food101 dataset provides images categorized into 101 food classes. For object detection with YOLO, you need not just the image but also bounding box annotations for each food item within the images. The standard Food101 dataset does not provide these bounding box annotations directly.

```python
Dataset: Food-Recognition-2022, https://www.kaggle.com/datasets/sainikhileshreddy/food-recognition-2022
```

Step-by-step methodology:

1. Download and Unpack the Dataset.
2. Convert Annotations from COCO Segmentation to YOLO Bounding Box Format.
3. Organize Data into YOLO Directory Structure.
4. Create a data.yaml Configuration File.
5. Install Ultralytics YOLOv8.
6. Train the YOLOv8 Model.
7. Evaluate and Infer.

```python
import zipfile
import os

# Define the path to your downloaded zip file
zip_file_path = 'food-recognition-2022.zip' # Assuming it's in the same directory as your notebook
# Or provide a full path: zip_file_path = 'C:/Users/YourUser/Downloads/food-recognition-2022.zip'

# Define the directory where you want to extract the contents
extract_dir = 'food_recognition_2022_dataset'

# Create the extraction directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

print(f"Attempting to unzip '{zip_file_path}' to '{extract_dir}'...")

try:
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extract_dir)
    print("Unzipping complete!")
except FileNotFoundError:
    print(f"Error: The zip file '{zip_file_path}' was not found. Please check the path.")
except Exception as e:
    print(f"An error occurred during unzipping: {e}")
```

```python
from ultralytics import YOLO

# Load a pre-trained YOLOv8 model (e.g., yolov8n.pt for nano, yolov8s.pt for small)
# Starting with 'n' (nano) or 's' (small) is good for initial runs due to the large dataset.
model = YOLO('yolov8n.pt')

# Train the model
# 'data': points to your food_detection.yaml
# 'epochs': Number of training iterations. Start with a reasonable number (e.g., 50-100)
#           and increase if the model is still improving. This dataset is large, so training
#           will take a long time.
# 'imgsz': Image size. 640 is a common choice. Adjust if you have memory issues or need to detect very small objects.
# 'batch': Batch size. Adjust based on your GPU memory. Smaller if you get out-of-memory errors.
# 'name': A unique name for your training run. Results will be saved in runs/detect/your_name
# 'device': Specify GPU device (e.g., 0 for single GPU, 0,1 for multiple, or 'cpu')
results = model.train(
    data='yolo_food_detection.yaml', # Path to your data config file
    epochs=50,                  # Number of training epochs
    imgsz=640,                   # Input image size
    batch=16,                    # Batch size (adjust based on GPU memory)
    name='yolov8_food_recognition', # Name for this training run
    device=0,                    # Use GPU 0. Change to 'cpu' if no GPU.
    # Optional: Add more parameters as needed
    # optimizer='AdamW',
    # lr0=0.001,
    patience=5 # Early stopping patience
)
```

```python
# Evaluate the model on the test dataset

from ultralytics import YOLO

model = YOLO('runs/detect/yolov8_food_recognition/weights/best.pt') # Path to your trained model

# Evaluate the model on the test dataset
metrics = model.val(data='yolo_food_detection.yaml', split='test', imgsz=640)

print(f"mAP50-95: {metrics.box.map}")
print(f"mAP50: {metrics.box.map50}")
print(f"mAP75: {metrics.box.map75}")
```

```python
# Test model with new image

from ultralytics import YOLO

model = YOLO('runs/detect/yolov8_food_recognition/weights/best.pt') # Path to your trained model

# Run inference on a specific image or folder
results = model.predict(source='path/to/your/new_food_image.jpg', conf=0.25, iou=0.7)

# Show results (if you have OpenCV installed, it will pop up windows)
for r in results:
    # r.boxes.xyxy # Bounding box coordinates (pixel)
    # r.boxes.conf # Confidence scores
    # r.boxes.cls  # Class IDs
    # r.names      # Class names mapping

    im_bgr = r.plot() # Plot detections on the image
    import cv2
    cv2.imshow('Food Detection', im_bgr)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    # Or save the image with detections
    # cv2.imwrite('detected_food_image.jpg', im_bgr)
```
